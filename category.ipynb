```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
us_data = pd.read_csv('data/youtube-new/USvideos.csv')
gb_data = pd.read_csv('data/youtube-new/GBvideos.csv')
# mx_data = pd.read_csv('data/youtube-new/MXvideos.csv')
# kr_data = pd.read_csv('data/youtube-new/KRvideos.csv')
# jp_data = pd.read_csv('data/youtube-new/JPvideos.csv')
# in_data = pd.read_csv('data/youtube-new/INvideos.csv')
fr_data = pd.read_csv('data/youtube-new/FRvideos.csv')
de_data = pd.read_csv('data/youtube-new/DEvideos.csv')
ca_data = pd.read_csv('data/youtube-new/CAvideos.csv')
```

```python
us_category_df_ND = us_data[['video_id', 'category_id']].drop_duplicates()
us_category_df = us_data[['video_id', 'category_id']]
grpby_obj = us_category_df.groupby('video_id').count()
grpby_obj.sort_values(by = 'category_id', ascending = False)
us_trends = pd.merge(grpby_obj.sort_values(by = 'category_id', ascending = False), us_category_df_ND, on = 'video_id' )
us_trends.info()
```

```python
gb_category_df_ND = gb_data[['video_id', 'category_id']].drop_duplicates()
gb_category_df = gb_data[['video_id', 'category_id']]
gb_grpby_obj = gb_category_df.groupby('video_id').count()
gb_grpby_obj.sort_values(by = 'category_id', ascending = False)
gb_trends = pd.merge(gb_grpby_obj.sort_values(by = 'category_id', ascending = False), gb_category_df_ND, on = 'video_id' )
gb_trends.head()
```

```python
fr_category_df_ND = fr_data[['video_id', 'category_id']].drop_duplicates()
fr_category_df = fr_data[['video_id', 'category_id']]
fr_grpby_obj = fr_category_df.groupby('video_id').count()
fr_grpby_obj.sort_values(by = 'category_id', ascending = False)
fr_trends = pd.merge(fr_grpby_obj.sort_values(by = 'category_id', ascending = False), fr_category_df_ND, on = 'video_id' )
fr_trends.head()
```

```python
de_category_df_ND = de_data[['video_id', 'category_id']].drop_duplicates()
de_category_df = de_data[['video_id', 'category_id']]
de_grpby_obj = de_category_df.groupby('video_id').count()
de_grpby_obj.sort_values(by = 'category_id', ascending = False)
de_trends = pd.merge(de_grpby_obj.sort_values(by = 'category_id', ascending = False), de_category_df_ND, on = 'video_id' )
de_trends.head()
```

```python
ca_category_df_ND = ca_data[['video_id', 'category_id']].drop_duplicates()
ca_category_df = ca_data[['video_id', 'category_id']]
ca_grpby_obj = ca_category_df.groupby('video_id').count()
ca_grpby_obj.sort_values(by = 'category_id', ascending = False)
ca_trends = pd.merge(ca_grpby_obj.sort_values(by = 'category_id', ascending = False), ca_category_df_ND, on = 'video_id' )
ca_trends.info()
```

```python
frame_list = [us_trends[['video_id']], gb_trends[['video_id']], fr_trends[['video_id']], de_trends[['video_id']], ca_trends[['video_id']]]
```

```python
sim_list = [[pd.merge(frame_list[i], frame_list[j], how='inner').count() for i in range(len(frame_list))] for j in range(len(frame_list))]
sim_list[0][1]['video_id']
sim_list = [[sim_list[i][j]['video_id'] for i in range(len(sim_list))] for j in range(len(sim_list))]

sim_list = np.array(sim_list)
for i in range(len(sim_list)):
    sim_list[i][i] = 0
sim_list
```

```python
countries = ["US", "Great Britain", "France", "Germany", "Canada"]

fig, ax = plt.subplots(figsize=(10,5))
im = ax.imshow(sim_list, cmap='YlOrRd')

ax.set_xticks(np.arange(len(countries)))
ax.set_yticks(np.arange(len(countries)))

ax.set_xticklabels(countries)
ax.set_yticklabels(countries)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(countries)):
    for j in range(len(countries)):
        text = ax.text(j, i, sim_list[i, j],
                       ha="center", va="center", color="b")
        
ax.set_title("Videos Trending in Multiple Countries")
fig.tight_layout()
plt.show()
plt.savefig('heatmap.jpg')
```
